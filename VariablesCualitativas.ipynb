# 0. Variables Cualitativas

Las **variables cualitativas** representan características o categorías que describen una población o conjunto de datos. A diferencia de las variables cuantitativas, estas no expresan cantidades numéricas, sino atributos o clasificaciones.

En este documento se trabajará con un conjunto de datos que representa el **nivel de estudios** de una población simulada, utilizando Python y la biblioteca **NumPy** para generar y organizar la información.

```python
# ============================================================
# Importación de bibliotecas
# ============================================================

# NumPy se utilizará para crear y manipular arreglos.
import numpy as np

# Random permitirá fijar una semilla para obtener resultados reproducibles.
import random

# Establecer una semilla para que los resultados sean siempre los mismos.
random.seed(42)
```

## Creación del conjunto de datos

En este ejemplo se construye una lista que representa el **nivel de estudios** de diferentes personas.

Cada categoría se repite un número determinado de veces para simular la cantidad de individuos pertenecientes a cada nivel educativo.

Posteriormente, los datos se mezclan aleatoriamente para evitar que queden ordenados por categorías.

```python
# ============================================================
# Crear el conjunto de datos.
# ============================================================

# Crear una lista vacía donde se almacenarán los niveles de estudio.
nivel_estudios = []

# Agregar 110 registros con el valor "Secundaria".
nivel_estudios.extend(["Secundaria"] * 110)

# Agregar 172 registros con el valor "Bachillerato".
nivel_estudios.extend(["Bachillerato"] * 172)

# Agregar 145 registros con el valor "Grado".
nivel_estudios.extend(["Grado"] * 145)

# Agregar 52 registros con el valor "Master".
nivel_estudios.extend(["Master"] * 52)

# Agregar 21 registros con el valor "Doctorado".
nivel_estudios.extend(["Doctorado"] * 21)

# Mezclar aleatoriamente los elementos de la lista.
nivel_estudios = np.random.permutation(nivel_estudios)

# Mostrar los primeros 10 elementos para verificar el resultado.
display(nivel_estudios[:10])
```

```python
print(nivel_estudios)
```

## Diagrama de Barras

El **diagrama de barras** es una representación gráfica utilizada para mostrar la frecuencia de cada categoría de una variable cualitativa.

En este ejemplo se calculará la cantidad de personas que pertenecen a cada **nivel de estudios** y posteriormente se representarán dichas frecuencias mediante un gráfico de barras.

Para ello se utilizará la función `np.unique()`, la cual permite obtener:

- Las categorías existentes.
- La frecuencia de aparición de cada categoría.

```python
# ============================================================
# Diagrama de barras
# ============================================================

# Importar la biblioteca para crear gráficos.
import matplotlib.pyplot as plt

# NumPy ya fue importado anteriormente.
import numpy as np

# Obtener las categorías únicas y la cantidad de veces
# que aparece cada una de ellas.
estudios, frecuencia = np.unique(
    nivel_estudios,
    return_counts=True
)

# Mostrar en pantalla las categorías y sus frecuencias.
print("Categorías:", estudios)
print("Frecuencias:", frecuencia)
```

```python
# ============================================================
# Crear el diagrama de barras.
# ============================================================

# Crear una figura con un tamaño adecuado para el PDF.
plt.figure(figsize=(8, 5))

# Etiqueta del eje Y.
plt.ylabel("Frecuencia absoluta")

# Etiqueta del eje X.
plt.xlabel("Nivel de estudios")

# Título del gráfico.
plt.title("Diagrama de barras")

# Crear el gráfico de barras.
plt.bar(
    estudios,
    frecuencia,
    color="steelblue",
    edgecolor="black"
)

# Agregar una cuadrícula horizontal para facilitar la lectura.
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Mostrar el gráfico.
plt.show()
```

> **Interpretación del gráfico**
>
> El diagrama de barras muestra la frecuencia absoluta de cada nivel de estudios presente en el conjunto de datos.
>
> Se observa que la categoría **Bachillerato** presenta la mayor cantidad de registros, mientras que **Doctorado** corresponde al grupo con menor frecuencia.
>
> Este tipo de representación facilita la comparación entre categorías y es una de las formas más utilizadas para visualizar variables cualitativas.

## Diagrama de Sectores

El **diagrama de sectores**, también conocido como **gráfico circular** o **Pie Chart**, representa la proporción que ocupa cada categoría respecto al total de los datos.

Cada sector del círculo corresponde a una categoría y su tamaño depende de la frecuencia de dicha categoría.

En este ejemplo se utilizarán las frecuencias obtenidas anteriormente para visualizar la distribución porcentual de los diferentes niveles de estudio.

```python
# ============================================================
# Crear un diagrama de sectores (Pie Chart).
# ============================================================

# Definir una lista de colores para cada categoría.
# Se utiliza una lista para conservar el orden de los colores.
colores = [
    "red",
    "blue",
    "orange",
    "green",
    "cyan"
]

# Crear una figura con un tamaño adecuado.
plt.figure(figsize=(7, 7))

# Crear el gráfico circular.
plt.pie(
    frecuencia,          # Valores que determinarán el tamaño de cada sector.
    labels=estudios,     # Etiquetas de cada categoría.
    colors=colores,      # Colores asignados a cada sector.
    autopct="%1.2f%%"    # Mostrar el porcentaje con dos decimales.
)

# Agregar un título.
plt.title("Diagrama de sectores")

# Mostrar el gráfico.
plt.show()
```

> **Interpretación del gráfico**
>
> El diagrama de sectores permite visualizar el porcentaje que representa cada nivel de estudios respecto al total de la población.
>
> Gracias a esta representación es posible identificar rápidamente cuáles son las categorías con mayor y menor participación.
>
> En este caso, **Bachillerato** ocupa el mayor porcentaje del conjunto de datos, mientras que **Doctorado** representa la menor proporción.

# Tarea: Implementación utilizando Pandas

En esta actividad se utilizará la biblioteca **Pandas** para representar la misma información obtenida anteriormente.

Pandas facilita el análisis de datos mediante estructuras como **Series** y **DataFrame**, además de incorporar métodos propios para generar gráficos utilizando internamente la biblioteca **Matplotlib**.

En esta práctica se representará:

- Diagrama de barras.
- Diagrama de sectores.

```python
# ============================================================
# Importar la biblioteca Pandas.
# ============================================================

import pandas as pd

# ============================================================
# Crear una Serie de Pandas con la frecuencia de cada nivel
# de estudios.
# ============================================================

serie_estudios = pd.Series(
    data=frecuencia,
    index=estudios,
    name="Frecuencia"
)

# Mostrar la información.
display(serie_estudios)
```

```python
# ============================================================
# Crear un diagrama de barras utilizando Pandas.
# ============================================================

# Crear el gráfico.
serie_estudios.plot(
    kind="bar",
    figsize=(8,5),
    color="steelblue",
    edgecolor="black",
    title="Diagrama de barras con Pandas"
)

# Etiquetas de los ejes.
plt.xlabel("Nivel de estudios")
plt.ylabel("Frecuencia absoluta")

# Agregar una cuadrícula horizontal.
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Mostrar el gráfico.
plt.show()
```

```python
# ============================================================
# Crear un diagrama de sectores utilizando Pandas.
# ============================================================

# Crear el gráfico circular.
serie_estudios.plot(
    kind="pie",
    figsize=(7,7),
    autopct="%1.2f%%",
    startangle=90,
    ylabel="",
    title="Diagrama de sectores con Pandas"
)

# Mostrar el gráfico.
plt.show()
```

```python
# ============================================================
# Crear un DataFrame con la información de los niveles de estudio.
# ============================================================

df_estudios = pd.DataFrame({
    "Nivel de estudios": estudios,
    "Frecuencia": frecuencia
})

# Mostrar el DataFrame.
display(df_estudios)
```

En esta práctica se trabajó con una variable cualitativa correspondiente al **nivel de estudios** de una población simulada. Se representó la información mediante diagramas de barras y de sectores utilizando tanto **Matplotlib** como **Pandas**.

El uso de Pandas simplifica considerablemente la generación de gráficos estadísticos, ya que integra de forma directa las funcionalidades de Matplotlib, permitiendo visualizar la información con menos código y de manera más intuitiva.
