# 1 Operaciones en Pandas

## Búsqueda

Pandas proporciona diversas herramientas para realizar búsquedas y consultas dentro de un **DataFrame**. Estas operaciones permiten localizar columnas, verificar la existencia de determinados valores y generar máscaras booleanas que facilitan el filtrado de la información.

Entre las operaciones más utilizadas se encuentran:

- Buscar columnas mediante su nombre.
- Comprobar si determinados valores existen dentro del DataFrame.
- Contar el número de ocurrencias de un valor específico.
- Crear máscaras booleanas para aplicar filtros sobre los datos.

```python
# ============================================================
# Importar las librerías necesarias
# ============================================================

import pandas as pd
import numpy as np
```

```python
# ============================================================
# Mostrar la versión instalada de Pandas
# ============================================================

pd.__version__
```

### Búsqueda en DataFrames

Pandas ofrece diferentes herramientas para buscar información dentro de un **DataFrame**. Es posible verificar si una columna existe, localizar determinados valores mediante máscaras booleanas y contar cuántas veces aparece un dato específico.

Las funciones `isin()` y las operaciones de comparación permiten generar máscaras que facilitan el filtrado y análisis de los datos.

```python
# ============================================================
# Crear un DataFrame con valores aleatorios
# ============================================================

# Generar una matriz de 2 filas y 3 columnas con números aleatorios
# entre 0 y 5.
rand_matrix = np.random.randint(6, size=(2, 3))

# Crear el DataFrame utilizando las columnas A, B y C.
frame = pd.DataFrame(rand_matrix, columns=list("ABC"))

# Mostrar el DataFrame.
display(frame)
```

```python
# ============================================================
# Buscar columnas dentro del DataFrame
# ============================================================

# Verificar si existe una columna llamada "A".
print("¿Existe la columna 'A'?", 'A' in frame)
```

### Búsqueda de valores

El método `isin()` permite comprobar si los valores del DataFrame pertenecen a un conjunto determinado. Como resultado se obtiene otra tabla con valores booleanos (`True` o `False`), conocida como **máscara**.

```python
# ============================================================
# Buscar valores específicos
# ============================================================

# Crear una máscara que indique cuáles elementos son 2 o 3.
display(frame.isin([2, 3]))
```

```python
# ============================================================
# Contar ocurrencias de un valor
# ============================================================

# Contar cuántas veces aparece el número 4.
print("Cantidad de valores iguales a 4:")
print(frame.isin([4]).values.sum())

# Mostrar la máscara correspondiente.
display(frame.isin([4]))

# Mostrar la máscara como un arreglo de NumPy.
print(frame.isin([4]).values)

# Mostrar el tipo del objeto retornado.
print(type(frame.isin([4]).values))
```

```python
# ============================================================
# Crear una máscara utilizando operadores relacionales
# ============================================================

# Seleccionar todos los valores mayores o iguales a 2.
mask = frame >= 2

# Contar cuántos valores cumplen la condición.
print("Cantidad de valores mayores o iguales a 2:")
print(mask.values.sum())

# Mostrar la máscara.
display(mask)
```

## Ordenación

Pandas permite ordenar fácilmente la información de un **DataFrame** mediante sus índices o los valores contenidos en las columnas. Estas operaciones facilitan el análisis y la organización de grandes conjuntos de datos.

```python
# ============================================================
# Crear un DataFrame desordenado
# ============================================================

from random import shuffle

# Generar una matriz de 5 filas y 4 columnas.
rand_matrix = np.random.randint(20, size=(5, 4))

# Crear una lista de índices.
indices = list(range(5))

# Mezclar los índices aleatoriamente.
shuffle(indices)

# Crear el DataFrame utilizando columnas con nombres personalizados.
frame = pd.DataFrame(rand_matrix,
                     columns=list("DACB"),
                     index=indices)

# Mostrar el DataFrame original.
display(frame)
```

```python
# ============================================================
# Ordenar por índice
# ============================================================

# Ordenar el DataFrame según el índice en orden descendente.
display(frame.sort_index(ascending=False))

# El DataFrame original permanece sin cambios.
display(frame)
```

```python
# ============================================================
# Ordenar las columnas
# ============================================================

# Ordenar las columnas alfabéticamente en orden descendente.
display(frame.sort_index(axis=1, ascending=False))
```

```python
# ============================================================
# Ordenar las filas utilizando una columna
# ============================================================

# Ordenar las filas según los valores de la columna A.
display(frame.sort_values(by='A', ascending=True))
```

```python
# ============================================================
# Ordenar las columnas utilizando los valores de una fila
# ============================================================

# Ordenar las columnas tomando como referencia los valores
# de la fila cuyo índice es 1.
display(frame.sort_values(by=1, axis=1, ascending=True))
```

```python
# ============================================================
# Ordenar el DataFrame y guardar los cambios
# ============================================================

# Ordenar las filas por la columna A en orden descendente.
# inplace=True modifica directamente el DataFrame original.
frame.sort_values(by='A', ascending=False, inplace=True)

# Mostrar el resultado final.
display(frame)
```

## Ranking

El **ranking** permite asignar una posición relativa a cada valor dentro de una Serie o DataFrame. Esta herramienta es útil para comparar datos, identificar los valores más altos o más bajos y establecer clasificaciones.

Pandas proporciona el método `rank()`, el cual admite diferentes estrategias para manejar valores repetidos mediante el parámetro `method`.

```python
# ============================================================
# Mostrar el DataFrame actual
# ============================================================

# Visualizar el DataFrame sobre el que se realizará el ranking.
display(frame)
```

```python
# ============================================================
# Construir un ranking por filas
# ============================================================

# Calcular el ranking de los valores de cada fila.
# method='max' asigna el rango más alto cuando existen empates.
# axis=1 indica que el cálculo se realiza horizontalmente (por filas).
display(frame.rank(method='max', axis=1))
```

### Ordenar valores de una columna

También es posible ordenar una columna de mayor a menor y recorrer sus valores mediante un ciclo `for`. Esto resulta útil cuando se desea procesar cada dato de forma individual después de haber sido ordenado.

```python
# ============================================================
# Recorrer una columna ordenada de mayor a menor
# ============================================================

# Ordenar el DataFrame utilizando la columna C en orden descendente.
# Posteriormente se obtiene únicamente dicha columna y se recorren
# sus valores uno por uno.

for valor in frame.sort_values(by='C', ascending=False)['C'].values:
    print(valor)
```

# 2 Operaciones

## Operaciones matemáticas entre objetos

Pandas permite realizar operaciones matemáticas entre **Series** y **DataFrames** de forma sencilla. Estas operaciones pueden efectuarse utilizando los operadores tradicionales (`+`, `-`, `*`, `/`) o mediante métodos específicos como `add()`, `sub()`, `mul()` y `div()`.

Cuando dos estructuras tienen las mismas dimensiones, las operaciones se realizan elemento por elemento. Si sus dimensiones son diferentes, Pandas alinea automáticamente filas y columnas utilizando sus índices y etiquetas.

```python
# ============================================================
# Crear dos DataFrames con valores aleatorios
# ============================================================

# Generar dos matrices de 4x4 con números aleatorios entre 0 y 99.
matrixA = np.random.randint(100, size=(4, 4))
matrixB = np.random.randint(100, size=(4, 4))

# Convertir las matrices en DataFrames.
frameA = pd.DataFrame(matrixA)
frameB = pd.DataFrame(matrixB)

# Mostrar ambos DataFrames.
display(frameA)
display(frameB)
```

```python
# ============================================================
# Suma mediante operadores y mediante métodos
# ============================================================

# Comparar si ambas formas de realizar la suma producen el mismo resultado.
display(frameA + frameB == frameA.add(frameB))

# Mostrar el resultado de la suma.
display(frameA + frameB)
```

```python
# ============================================================
# Resta mediante operadores y mediante métodos
# ============================================================

# Comparar si ambas formas de realizar la resta producen el mismo resultado.
display(frameB - frameA == frameB.sub(frameA))

# Mostrar el resultado de la resta.
display(frameB - frameA)
```

### Operaciones entre DataFrames con diferentes dimensiones

Cuando dos DataFrames no poseen el mismo número de filas o columnas, Pandas intenta alinear automáticamente los datos utilizando los índices y nombres de las columnas. Los elementos que no tienen correspondencia se representan con el valor **NaN**.

```python
# ============================================================
# Operaciones entre DataFrames de distinto tamaño
# ============================================================

# Crear un DataFrame de 3x3.
frameC = pd.DataFrame(np.random.randint(100, size=(3, 3)))

# Mostrar ambos DataFrames.
display(frameA)
display(frameC)

# Realizar la suma.
# Los elementos sin correspondencia aparecerán como NaN.
display(frameC + frameA)
```

```python
# ============================================================
# Utilizar un valor por defecto en operaciones
# ============================================================

# fill_value=0 reemplaza los valores inexistentes por cero
# antes de realizar la operación.
display(frameA.add(frameC, fill_value=0))
```

### Operadores aritméticos y tipos de datos

Las operaciones matemáticas únicamente pueden realizarse sobre datos numéricos. Si un DataFrame contiene cadenas de texto u otros tipos incompatibles, Pandas generará una excepción indicando que la operación no es válida.

```python
# ============================================================
# Ejemplo de operación no válida
# ============================================================

# Crear un DataFrame con valores de tipo texto.
frameD = pd.DataFrame({
    0: ['a', 'b'],
    1: ['d', 'f']
})

display(frameD)

# Intentar realizar una resta entre un DataFrame numérico
# y otro de texto producirá un error.
# Se utiliza try-except para mostrar el mensaje sin detener
# la ejecución del Notebook.

try:
    display(frameA - frameD)
except Exception as e:
    print("Error:", e)
```

## Operaciones entre Series y DataFrames

Las filas y columnas de un DataFrame son objetos de tipo **Series**. Gracias a ello, es posible realizar operaciones matemáticas entre una Serie y un DataFrame completo. Pandas se encarga de alinear automáticamente los datos utilizando sus índices y nombres de columnas.

```python
# ============================================================
# Crear un DataFrame para operaciones con Series
# ============================================================

# Generar una matriz de 3 filas y 4 columnas.
rand_matrix = np.random.randint(10, size=(3, 4))

# Crear el DataFrame con nombres de columnas.
df = pd.DataFrame(rand_matrix, columns=list("ABCD"))

# Mostrar el DataFrame.
display(df)

# Mostrar la primera fila como una Serie.
display(df.iloc[0])

# Mostrar el tipo del objeto.
display(type(df.iloc[0]))
```

```python
# ============================================================
# Restar una Serie a todas las filas del DataFrame
# ============================================================

# Restar la primera fila al resto del DataFrame.
display(df - df.iloc[0])

# La operación anterior también puede realizarse mediante
# el método sub().
display(df.sub(df.iloc[0], axis=1))
```

```python
# ============================================================
# Restar una columna a todas las filas
# ============================================================

# Restar los valores de la columna A a todas las columnas
# del DataFrame.
display(df.sub(df["A"], axis=0))
```

### Aplicación de funciones a medida con lambda

```python
# ============================================================
# Crear un DataFrame con valores enteros aleatorios.
# ============================================================

rand_matrix = np.random.randint(10, size=(3, 4))

frame = pd.DataFrame(rand_matrix, columns=list("ABCD"))

display(frame)

# Aplicar una función lambda a cada fila (axis=1).
# La función calcula la diferencia entre el valor máximo y el mínimo.
print(frame.apply(lambda x: x.max() - x.min(), axis=1))
```

```python
# ============================================================
# Definir una función que calcule la diferencia entre
# el valor máximo y el mínimo de una Serie.
# ============================================================

def max_min(x):
    return x.max() - x.min()

# Aplicar la función a cada columna (axis=0).
print(frame.apply(max_min, axis=0))
```

```python
# ============================================================
# Crear un nuevo DataFrame para calcular la diferencia
# entre el valor máximo y mínimo de cada fila.
# ============================================================

rand_matrix = np.random.randint(10, size=(3, 4))

frame = pd.DataFrame(rand_matrix, columns=list("ABCD"))

display(frame)

# Aplicar una función lambda sobre las filas (axis=1).
# Devuelve la diferencia entre el valor máximo y el mínimo de cada fila.
print(frame.apply(lambda x: x.max() - x.min(), axis=1))
```

# 3 Estadística descriptiva

- Permite realizar un análisis preliminar de los datos.
- Disponible tanto para **Series** como para **DataFrames**.

| Operación | Descripción |
|:----------|:------------|
| `count` | Número de valores distintos de NaN. |
| `describe` | Muestra un resumen estadístico de los datos. |
| `min`, `max` | Valor mínimo y máximo. |
| `argmin`, `argmax` | Índices posicionales del valor mínimo y máximo. |
| `idxmin`, `idxmax` | Índices semánticos del valor mínimo y máximo. |
| `sum` | Suma de los elementos. |
| `mean` | Media aritmética de los elementos. |
| `median` | Mediana de los elementos. |
| `mad` | Desviación absoluta media. |
| `var` | Varianza de los elementos. |
| `std` | Desviación estándar de los elementos. |
| `cumsum` | Suma acumulada de los elementos. |
| `diff` | Diferencia entre elementos consecutivos. |

```python
# ============================================================
# Crear un DataFrame con diferentes tipos de datos
# y obtener un resumen estadístico.
# ============================================================

diccionario = {
    "nombre": ["Marisa", "Laura", "Manuel", "Carlos"],
    "edad": [34, 34, 11, 30],
    "puntos": [98, 12, 98, np.nan],
    "genero": ["F", "F", "M", "M"]
}

frame = pd.DataFrame(diccionario)

display(frame)

# Mostrar un resumen estadístico de las columnas numéricas.
display(frame.describe())
```

```python
# ============================================================
# Aplicar operaciones básicas sobre el DataFrame.
# ============================================================

# Suma de cada columna.
print(frame.sum())

display(frame)

# Suma de los valores numéricos por fila.
print(frame.sum(axis=1, numeric_only=True))
```

```python
# ============================================================
# Calcular la media de las columnas numéricas.
# ============================================================

display(frame.mean(numeric_only=True))
```

```python
# ============================================================
# Calcular la suma acumulada de cada columna.
# ============================================================

display(frame.cumsum())
```

```python
# ============================================================
# Contar la cantidad de valores no nulos por fila.
# ============================================================

display(frame.count(axis=1))
```

```python
# ============================================================
# Calcular la desviación estándar de la columna "edad".
# ============================================================

print(frame["edad"].std())
```

```python
# ============================================================
# Obtener el índice del valor máximo de la columna "edad".
# ============================================================

print(frame["edad"].idxmax())
```

```python
# ============================================================
# Obtener el índice del valor mínimo de la columna "puntos".
# ============================================================

print(frame["puntos"].idxmin())
```

```python
# ============================================================
# Mostrar las filas que contienen el valor máximo
# de la columna "puntos".
# ============================================================

print(frame["puntos"].max())

display(frame[frame["puntos"] == frame["puntos"].max()])
```

```python
# ============================================================
# Crear una nueva columna con el ranking de los puntajes.
# ============================================================

frame["ranking"] = frame["puntos"].rank(method="max")
```

```python
# ============================================================
# Mostrar el DataFrame actualizado con la columna ranking.
# ============================================================

display(frame)
```

### 3.1 Agregaciones

Las **agregaciones** permiten resumir la información de un conjunto de datos agrupándolos por una o varias columnas. Entre las operaciones más comunes se encuentran el cálculo del promedio, mínimo, máximo y conteo de registros.

```python
# ============================================================
# Agrupar los datos por el género y contar los valores
# de cada columna dentro de cada grupo.
# ============================================================

display(frame)

df = frame.groupby("genero").count()

display(df)
```

```python
# ============================================================
# Agrupar por la columna "puntos".
# Los valores NaN no forman parte de ningún grupo.
# ============================================================

df = frame.groupby("puntos").count()

display(df)
```

```python
# ============================================================
# Calcular el promedio de las columnas numéricas
# agrupando por género.
# ============================================================

display(frame.groupby("genero").mean(numeric_only=True))
```

```python
# ============================================================
# Obtener el valor máximo de cada columna
# agrupando por género.
# ============================================================

display(frame.groupby("genero").max())
```

```python
# ============================================================
# Aplicar varias funciones estadísticas a múltiples columnas.
# ============================================================

display(
    frame.groupby("genero")[["edad", "puntos"]]
    .aggregate(["min", "mean", "max"])
)
```

```python
# ============================================================
# Definir una función para filtrar únicamente aquellos grupos
# cuya edad promedio sea mayor a 30 años.
# ============================================================

def media(x):
    return x["edad"].mean() > 30

display(frame)

display(frame.groupby("genero").filter(media))
```

### 3.2 Correlaciones

Las **correlaciones** permiten medir la relación existente entre dos variables.

Su valor puede variar entre:

- **1** → Relación positiva perfecta.
- **0** → No existe relación lineal.
- **-1** → Relación negativa perfecta.

Pandas proporciona el método `corr()` para calcular el coeficiente de correlación entre columnas numéricas.

#### 3.2.1 Ejemplo: Fuel Efficiency

Se utilizará el conjunto de datos **Auto MPG**, disponible en el repositorio de la Universidad de California (UCI), para analizar la relación entre distintas variables de vehículos.

Fuente:
https://archive.ics.uci.edu/ml/datasets/Auto+MPG

```python
# ============================================================
# Importar el conjunto de datos Auto MPG desde la UCI.
# ============================================================

import pandas as pd

path = (
    "http://archive.ics.uci.edu/ml/machine-learning-databases/"
    "auto-mpg/auto-mpg.data"
)

mpg_data = pd.read_csv(
    path,
    sep=r"\s+",
    header=None,
    names=[
        "mpg",
        "cilindros",
        "desplazamiento",
        "potencia",
        "peso",
        "aceleracion",
        "año",
        "origen",
        "nombre",
    ],
    na_values="?",
    engine="c",
)
```

```python
# ============================================================
# Mostrar una muestra aleatoria de cinco registros.
# ============================================================

display(mpg_data.sample(5))
```

```python
# ============================================================
# Mostrar un resumen estadístico del conjunto de datos.
# ============================================================

display(mpg_data.describe(include="all"))
```

```python
# ============================================================
# Mostrar información general del DataFrame.
# ============================================================

mpg_data.info()
```

#### 3.2.2 Correlaciones entre variables

```python
# ============================================================
# Calcular la correlación entre el rendimiento del vehículo
# (mpg) y su peso.
# ============================================================

print(mpg_data["mpg"].corr(mpg_data["peso"]))
```

```python
# ============================================================
# Calcular la correlación entre el peso del vehículo
# y su aceleración.
# ============================================================

print(mpg_data["peso"].corr(mpg_data["aceleracion"]))
```

### 3.2.3 Correlaciones entre todas las variables

Además de calcular la correlación entre dos columnas, Pandas permite obtener una **matriz de correlación**, la cual muestra el grado de relación entre todas las variables numéricas del conjunto de datos.

```python
# ============================================================
# Calcular la matriz de correlación entre todas las
# columnas numéricas del DataFrame.
# ============================================================

display(mpg_data.corr(numeric_only=True))
```

```python
# ============================================================
# Eliminar las columnas "año" y "origen" antes de calcular
# la correlación, ya que no son relevantes para este análisis.
# ============================================================

corr_data = mpg_data.drop(
    ["año", "origen"],
    axis=1
).corr(numeric_only=True)

display(corr_data)
```

```python
# ============================================================
# Importar Matplotlib para representar la matriz
# de correlación mediante colores.
# ============================================================

import matplotlib.pyplot as plt
```

```python
# ============================================================
# Mostrar la matriz de correlación utilizando un gradiente
# de colores para facilitar su interpretación.
# ============================================================

corr_data.style.background_gradient(
    cmap=plt.get_cmap("RdYlGn"),
    axis=1
)
```

```python
# ============================================================
# Obtener la variable con la correlación más negativa
# respecto a cada columna.
# ============================================================

display(
    mpg_data
    .drop(["año", "origen"], axis=1)
    .corr(numeric_only=True)
    .idxmin()
)
```

```python
# ============================================================
# Obtener la variable con la correlación más positiva.
# El resultado incluye la correlación de cada variable consigo misma.
# ============================================================

display(
    mpg_data
    .drop(["año", "origen"], axis=1)
    .corr(numeric_only=True)
    .idxmax()
)
```

```python
# ============================================================
# Crear una copia de la matriz de correlación y reemplazar
# la diagonal principal por ceros.
# ============================================================

positive_corr = (
    mpg_data
    .drop(["año", "origen"], axis=1)
    .corr(numeric_only=True)
)

# Crear una copia modificable del arreglo.
corr = positive_corr.to_numpy().copy()

# Reemplazar la diagonal principal por ceros.
np.fill_diagonal(corr, 0)

# Volver a crear el DataFrame con los mismos índices y columnas.
positive_corr = pd.DataFrame(
    corr,
    index=positive_corr.index,
    columns=positive_corr.columns
)

display(positive_corr)

display(positive_corr.idxmax())
```

```python
# ============================================================
# Aplicar un mapa de colores para visualizar las
# correlaciones positivas y negativas.
# ============================================================

positive_corr.style.background_gradient(
    cmap=plt.get_cmap("RdYlGn"),
    axis=1,
    vmin=-1.0,
    vmax=1.0
)
```
