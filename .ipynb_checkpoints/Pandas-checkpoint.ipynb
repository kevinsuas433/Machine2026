# 1 Pandas

- **Pandas** es la librería estándar en Python para trabajar con datos tabulares de forma eficiente.
- Permite almacenar diferentes tipos de datos en una misma estructura (texto, enteros, decimales, valores booleanos, entre otros).
- Sus dos estructuras principales son:

  - **Series:** estructura unidimensional.
  - **DataFrame:** estructura bidimensional o multidimensional, similar a una hoja de cálculo.

# 2 Series

Las **Series** son estructuras de datos unidimensionales proporcionadas por Pandas.

Sus principales características son:

- Almacenan una única columna de datos.
- Cada elemento posee un índice asociado.
- Tanto los datos como los índices pueden modificarse.
- Son similares a un arreglo de NumPy, pero ofrecen un manejo más flexible de los índices.

## 2.1 Creación de Series

Una **Serie (Series)** es una estructura unidimensional de Pandas que almacena una colección de datos junto con un índice.

Por defecto, Pandas asigna índices numéricos consecutivos comenzando desde **0**, aunque es posible personalizar dichos índices para facilitar la identificación de los datos.

````python
# ============================================================
# Crear una Serie utilizando los índices predeterminados
# ============================================================

# Se crea una Serie con varios países.
countries = pd.Series([
    'Spain',
    'Andorra',
    'Gibraltar',
    'Portugal',
    'France'
])

# Mostrar la Serie completa.
print(countries)
````

## Series con índices personalizados

Además de utilizar índices automáticos, Pandas permite definir índices personalizados, lo que facilita localizar la información utilizando identificadores específicos.

````python
# ============================================================
# Crear una Serie con índices personalizados
# ============================================================

# Se crea una Serie indicando manualmente los índices.
countries = pd.Series(
    ['Spain', 'Andorra', 'Gibraltar', 'Portugal', 'France'],
    index=range(10, 60, 10)
)

# Mostrar la Serie.
print(countries)
````

## Series con índices de tipo texto

Los índices no necesariamente deben ser números. También pueden ser cadenas de texto, caracteres u otros tipos de datos.

````python
# ============================================================
# Crear una Serie utilizando índices de texto
# ============================================================

# Se crea una Serie donde los índices son letras.
football_cities = pd.Series(
    ['Barcelona', 'Madrid', 'Valencia', 'Sevilla'],
    index=['a', 'b', 'c', 'd']
)

# Mostrar la Serie.
print(football_cities)
````

## Asignación de atributos

Las Series pueden tener un nombre descriptivo y también es posible asignar un nombre a su índice para facilitar la interpretación de la información.

````python
# ============================================================
# Asignar nombres a la Serie y a sus índices
# ============================================================

# Asignar un nombre descriptivo a la Serie.
football_cities.name = "Ciudades con dos equipos en primera"

# Asignar un nombre al índice.
football_cities.index.name = "Id"

# Mostrar la Serie con sus nuevos atributos.
print(football_cities)
````

## Acceso a los elementos de una Serie

Los elementos de una Serie pueden recuperarse utilizando:

- La posición del elemento.
- El valor del índice asignado.

Ambos métodos permiten acceder al mismo dato cuando corresponden al mismo elemento.

````python
# ============================================================
# Acceder a los elementos de una Serie
# ============================================================

# Acceso utilizando la posición (tercer elemento).
print(football_cities.iloc[2])

# Acceso utilizando el índice semántico.
print(football_cities.loc['c'])

# Comprobar que ambos accesos devuelven el mismo valor.
print(football_cities.iloc[2] == football_cities.loc['c'])
````

# 3 Tratamiento similar a `ndarray`

Las **Series** de Pandas comparten muchas características con los arreglos de **NumPy**. Es posible seleccionar varios elementos, realizar operaciones de slicing, convertir una Serie a otros tipos de datos y aplicar funciones de NumPy sobre ella.

## Selección de múltiples elementos

Una Serie permite recuperar varios elementos simultáneamente indicando una lista con los índices o posiciones que se desean obtener.

````python
# ============================================================
# Seleccionar múltiples elementos de una Serie
# ============================================================

# Selección utilizando los índices de la Serie.
print(football_cities[['a', 'c']])

# Selección utilizando posiciones mediante iloc.
print(football_cities.iloc[[0, 3]])
````

## Slicing

Al igual que las listas de Python, las Series permiten obtener subconjuntos mediante **slicing**.

Cuando se utilizan índices semánticos con `.loc`, el extremo final también se incluye.

````python
# ============================================================
# Ejemplos de slicing en Series
# ============================================================

# Slicing utilizando índices semánticos.
print(football_cities.loc[:'c'])

# Slicing utilizando posiciones.
print(football_cities.iloc[:2])
````

## Conversión de una Serie a lista

Una Serie puede convertirse fácilmente en una lista de Python utilizando la función `list()`.

````python
# ============================================================
# Convertir una Serie a lista
# ============================================================

lista = list(football_cities.loc[:'c'])

print(lista)
print(type(lista))
````

## Tipo del objeto resultante

Aunque se realice un slicing, el resultado continúa siendo una **Serie** de Pandas.

````python
# ============================================================
# Comprobar el tipo de dato
# ============================================================

print(type(football_cities.loc[:'c']))
````

## Conversión a un arreglo de NumPy

Una Serie puede transformarse en un arreglo (`ndarray`) para aprovechar todas las funciones de NumPy.

````python
# ============================================================
# Convertir una Serie a ndarray
# ============================================================

import numpy as np

cities = np.array(football_cities.loc[:'c'])

print(cities)
print(type(cities))
````

## Conversión a diccionario

También es posible convertir una Serie en un diccionario donde los índices se convierten en las claves y los valores permanecen asociados.

````python
# ============================================================
# Convertir una Serie a diccionario
# ============================================================

diccionario = dict(football_cities.loc[:'c'])

print(diccionario)
````

## Selección mediante máscaras

Las máscaras permiten filtrar únicamente aquellos elementos que cumplen una determinada condición lógica.

````python
# ============================================================
# Filtrado utilizando máscaras booleanas
# ============================================================

# Crear una Serie con la sucesión de Fibonacci.
fibonacci = pd.Series([0, 1, 1, 2, 3, 5, 8, 13, 21])

print(fibonacci)

# Crear una máscara con los valores mayores que 10.
mask = fibonacci > 10

print(mask)

# Mostrar únicamente los elementos que cumplen la condición.
print(fibonacci[mask])

# Serie esperada para realizar la comparación.
dst = pd.Series([13, 21])

print(dst)

# Comparación inicial (devuelve False porque los índices son diferentes).
print(dst.equals(fibonacci))

# Reiniciar los índices de la Serie filtrada.
fb = fibonacci[mask]
fb.reset_index(drop=True, inplace=True)

print(fb)

# Ahora ambas Series son iguales.
print(dst.equals(fb))
````

## Aplicación de funciones de NumPy

Las funciones de NumPy pueden utilizarse directamente sobre una Serie de Pandas.

````python
# ============================================================
# Aplicar funciones de NumPy sobre una Serie
# ============================================================

import numpy as np

print(np.sum(fibonacci))
````

## Filtrado utilizando `np.where`

La función `np.where()` permite reemplazar valores de una Serie dependiendo de si cumplen o no una determinada condición.

````python
# ============================================================
# Reemplazar valores nulos utilizando np.where
# ============================================================

distances = pd.Series([12.1, np.nan, 12.8, 76.9, 6.1, 7.2])

# Sustituir los valores NaN por el número 2.
valid_distances = np.where(pd.notnull(distances), distances, 2)

print(valid_distances)
print(type(valid_distances))
````

# 3.0.1 Iteración

Las **Series** de Pandas permiten recorrer fácilmente tanto sus valores como sus índices.

Existen varias formas de iterar sobre una Serie:

- Recorrer únicamente los valores.
- Recorrer únicamente los índices.
- Recorrer índices y valores simultáneamente mediante `items()`.
- Utilizar `zip()` para combinar índices y valores durante la iteración.

Estas opciones son muy útiles cuando se necesita procesar o mostrar la información almacenada en una Serie.

````python
# ============================================================
# Iterar sobre una Serie de Pandas
# ============================================================

import pandas as pd

# Crear una Serie con la sucesión de Fibonacci.
fibonacci = pd.Series([0, 1, 1, 2, 3, 5, 8, 13, 21])

# ------------------------------------------------------------
# Recorrer únicamente los valores de la Serie.
# ------------------------------------------------------------
print("Recorrido de valores:")

for value in fibonacci:
    print("Valor:", value)

print()

# ------------------------------------------------------------
# Recorrer únicamente los índices de la Serie.
# ------------------------------------------------------------
print("Recorrido de índices:")

for index in fibonacci.index:
    print("Índice:", index)
````

````python
# ============================================================
# Recorrer índices y valores al mismo tiempo utilizando items()
# ============================================================

for index, value in fibonacci.items():
    print(f"Índice: {index}  Valor: {value}")
````

````python
# ============================================================
# Otra forma de recorrer índices y valores utilizando zip()
# ============================================================

for index, value in zip(fibonacci.index, fibonacci):
    print(f"Índice: {index}  Valor: {value}")
````

# 3.1 Series como diccionarios

Una **Serie** de Pandas puede comportarse de manera muy similar a un diccionario de Python.

En este caso:

- Los índices actúan como las claves.
- Los valores almacenan la información asociada.
- Es posible agregar, modificar y eliminar elementos utilizando sus índices.
- También es posible consultar si una clave existe antes de acceder a ella para evitar errores.

````python
# ============================================================
# Crear una Serie a partir de un diccionario
# ============================================================

import pandas as pd

# Cada clave del diccionario será un índice de la Serie.
serie = pd.Series({
    "Carlos": 100,
    "Marcos": 98
})

# Mostrar información de la Serie.
print("Índices:")
print(serie.index)

print("\nValores:")
print(serie.values)

print("\nSerie completa:")
print(serie)

print("\nTipo del objeto:")
print(type(serie))
````

````python
# ============================================================
# Modificar una Serie como si fuera un diccionario
# ============================================================

# Agregar un nuevo elemento.
serie["Pedro"] = 12

# Modificar el valor existente.
serie["Pedro"] = 15

# Eliminar un elemento mediante su índice.
del serie["Marcos"]

print("Serie modificada:")
print(serie)
````

````python
# ============================================================
# Verificar si un índice existe antes de acceder a él
# ============================================================

# Evita errores si el índice no existe.
if "Marcos" in serie:
    print(serie["Marcos"])
else:
    print("El índice 'Marcos' no existe.")

print("\nSerie actual:")
print(serie)
````

# 3.2 Operaciones entre Series

Las **Series** de Pandas permiten realizar operaciones matemáticas de forma automática entre sus elementos.

Cuando dos Series poseen los mismos índices, las operaciones se realizan elemento por elemento.

Si alguno de los índices no existe en ambas Series, el resultado para ese índice será **NaN** (*Not a Number*), indicando que no fue posible realizar la operación.

Las operaciones más comunes son la suma, resta, multiplicación y división entre Series.

````python
# ============================================================
# Suma de dos Series
# ============================================================

import pandas as pd

# Crear la primera Serie.
serie1 = pd.Series([10, 20, 30, 40], index=range(4))

# Crear la segunda Serie.
serie2 = pd.Series([1, 2, 3], index=range(3))

# La suma se realiza utilizando los índices.
# Si un índice no existe en ambas Series, el resultado será NaN.
suma = serie1 + serie2

print("Serie 1:")
print(serie1)

print("\nSerie 2:")
print(serie2)

print("\nResultado de la suma:")
print(suma)
````

````python
# ============================================================
# Resta entre dos Series
# ============================================================

# La resta funciona igual que la suma:
# Pandas alinea automáticamente los elementos por su índice.
resta = serie1 - serie2

print("Resultado de la resta:")
print(resta)
````

````python
# ============================================================
# Reemplazar valores NaN después de una operación
# ============================================================

# Realizar nuevamente la suma.
resultado = serie1 + serie2

print("Resultado original:")
print(resultado)

# Identificar los elementos cuyo valor es NaN.
resultado[pd.isnull(resultado)] = 0

print("\nResultado después de reemplazar NaN por 0:")
print(resultado)
````

## 3.2.1 Diferencias entre Pandas Series y un diccionario

Aunque una **Serie** puede parecer similar a un diccionario de Python, existen diferencias importantes:

- Un **diccionario** relaciona claves y valores sin mantener una estructura posicional estricta.
- Una **Serie** organiza los datos mediante índices alineados con cada valor.
- Las Series permiten realizar operaciones matemáticas de forma automática entre elementos con el mismo índice.
- Los valores pueden almacenarse a partir de listas o arreglos de NumPy.
- Tanto los índices como los valores pueden ser numéricos o de texto.
- Las Series combinan características de las listas y los diccionarios, proporcionando una estructura eficiente para el análisis de datos en una dimensión.

# 4 DataFrame

Los **DataFrames** son la estructura de datos más utilizada en Pandas para trabajar con información tabular.

Un DataFrame está compuesto por filas y columnas, donde cada columna es una **Serie** que comparte el mismo índice con las demás.

Los DataFrames permiten almacenar diferentes tipos de datos (texto, enteros, decimales, booleanos, etc.) y facilitan su manipulación y análisis.

````python
# ============================================================
# Crear un DataFrame a partir de un diccionario
# ============================================================

import pandas as pd

# Crear un diccionario donde cada clave representa una columna.
diccionario = {
    "Nombre": ["Marisa", "Laura", "Manuel"],
    "Edad": [34, 29, 12]
}

print("Diccionario original:")
print(diccionario)

# Crear el DataFrame.
frame = pd.DataFrame(diccionario)

print("\nDataFrame generado:")
display(frame)
````

````python
# ============================================================
# Crear un DataFrame indicando índices personalizados
# ============================================================

# Los índices identifican cada fila.
frame = pd.DataFrame(
    diccionario,
    index=["a", "b", "c"]
)

display(frame)
````

````python
# ============================================================
# Especificar el orden de las columnas
# ============================================================

# También es posible indicar columnas que aún no existen.
# En ese caso Pandas rellenará esos valores con NaN.
frame = pd.DataFrame(
    diccionario,
    columns=[
        "Nacionalidad",
        "Nombre",
        "Edad",
        "Profesion",
        "Genero"
    ]
)

display(frame)
````

## Acceso a columnas

Cada columna de un DataFrame puede recuperarse como una **Serie**.

Existen dos formas principales de acceder a una columna:

- Utilizando corchetes (`frame["Nombre"]`), que es la forma recomendada.
- Utilizando la notación por punto (`frame.Nombre`), siempre que el nombre de la columna sea un identificador válido.

````python
# ============================================================
# Acceder a columnas del DataFrame
# ============================================================

# Obtener la columna "Nombre".
nombres = frame["Nombre"]

print("Columna Nombre:")
display(nombres)

print(type(nombres))

# Obtener la columna "Edad".
edades = frame["Edad"]

print("\nColumna Edad:")
display(edades)

print(type(edades))
````

````python
# ============================================================
# Acceso utilizando la notación por punto
# ============================================================

# Esta forma funciona únicamente cuando el nombre de la columna
# es un identificador válido.

profesiones = frame.Profesion

display(profesiones)

print(type(profesiones))
````

````python
# ============================================================
# Acceder a un valor específico del DataFrame
# ============================================================

# Obtener el primer nombre utilizando corchetes.
print(frame["Nombre"][0])

# Obtener el mismo valor utilizando la notación por punto.
print(frame.Nombre[0])

# Acceder al primer elemento de la Serie almacenada en la variable.
print(profesiones[0])
````

## 4.0.1 Formas de crear un DataFrame

Pandas ofrece varias maneras de crear un **DataFrame**, dependiendo del origen de los datos.

Las formas más comunes son:

- A partir de una **Serie** de Pandas.
- A partir de una **lista de diccionarios**.
- A partir de un **diccionario de Series**.
- A partir de un **array bidimensional de NumPy**.
- A partir de un **array estructurado de NumPy**.

La elección depende de cómo estén organizados los datos antes de ser procesados.

# 4.1 Modificar DataFrames

Los DataFrames son estructuras dinámicas que permiten agregar, modificar o eliminar tanto filas como columnas.

Entre las operaciones más comunes se encuentran:

- Agregar nuevas columnas.
- Modificar el contenido de una columna existente.
- Insertar nuevas filas.
- Eliminar filas o columnas.
- Obtener la transpuesta del DataFrame.

````python
# ============================================================
# Agregar nuevas columnas a un DataFrame
# ============================================================

import pandas as pd

# Crear el diccionario base.
diccionario = {
    "Nombre": ["Marisa", "Laura", "Manuel"],
    "Edad": [34, 29, 12]
}

# Crear el DataFrame indicando todas las columnas.
frame = pd.DataFrame(
    diccionario,
    columns=[
        "Nacionalidad",
        "Nombre",
        "Edad",
        "Profesion",
        "Direccion"
    ]
)

# Agregar una columna con el mismo valor para todas las filas.
frame["Direccion"] = "Desconocida"

display(frame)
````

````python
# ============================================================
# Asignar distintos valores a una columna
# ============================================================

# Lista con una dirección para cada persona.
lista_direcciones = [
    "Rue 13 del Percebe, 13",
    "Evergreen Terrace, 3",
    "Av. de los Rombos, 12"
]

# Reemplazar la columna con los nuevos valores.
frame["Direccion"] = lista_direcciones

display(frame)
````

````python
# ============================================================
# Agregar una nueva fila utilizando loc
# ============================================================

# Cada elemento corresponde a una columna del DataFrame.
usuario_nuevo = [
    "Alemania",
    "Klaus",
    20,
    "None",
    "Desconocida"
]

# Insertar la nueva fila en la posición 3.
frame.loc[3] = usuario_nuevo

display(frame)
````

````python
# ============================================================
# Eliminar filas y columnas
# ============================================================

# Crear nuevamente el DataFrame.
frame = pd.DataFrame(
    diccionario,
    columns=[
        "Nacionalidad",
        "Nombre",
        "Edad",
        "Profesion"
    ]
)

# ------------------------------------------------------------
# Eliminar una fila.
# drop() devuelve un nuevo DataFrame, por ello es necesario
# guardar nuevamente el resultado.
# ------------------------------------------------------------
frame = frame.drop(2)

display(frame)

# ------------------------------------------------------------
# Eliminar una columna.
# inplace=True modifica directamente el DataFrame.
# ------------------------------------------------------------
frame.drop("Nombre", axis=1, inplace=True)

display(frame)
````

````python
# ============================================================
# Eliminar una columna utilizando del
# ============================================================

# Eliminar completamente la columna "Profesion".
del frame["Profesion"]

display(frame)
````

````python
# ============================================================
# Mostrar la transpuesta del DataFrame
# ============================================================

# La propiedad T intercambia filas por columnas.
display(frame.T)
````

### Observaciones

- Para **agregar columnas**, basta con asignar un nuevo nombre de columna.
- Para **agregar filas**, se utiliza generalmente `loc`.
- El método **drop()** devuelve un nuevo DataFrame; por ello normalmente se reasigna el resultado o se utiliza `inplace=True`.
- La palabra clave **del** elimina definitivamente una columna.
- La propiedad **T** permite obtener la transpuesta del DataFrame, intercambiando filas por columnas.

# 4.2 Iteración

Los DataFrames permiten recorrer tanto sus columnas como sus filas.

Existen varias formas de iterar sobre un DataFrame:

- Recorrer los nombres de las columnas.
- Recorrer las filas mediante la propiedad `values`.
- Recorrer cada elemento individual utilizando bucles anidados.

La forma de iterar dependerá de la información que se necesite procesar.

````python
# ============================================================
# Iterar sobre las columnas de un DataFrame
# ============================================================

import pandas as pd

# Crear un diccionario con los datos.
diccionario = {
    "Nombre": ["Marisa", "Laura", "Manuel"],
    "Edad": [34, 29, 12]
}

# Crear el DataFrame.
frame = pd.DataFrame(
    diccionario,
    columns=["Nacionalidad", "Nombre", "Edad", "Profesion"]
)

display(frame)

# ------------------------------------------------------------
# Al recorrer directamente un DataFrame se obtienen los nombres
# de las columnas.
# ------------------------------------------------------------
for columna in frame:
    print(columna)

print(type(columna))
````

````python
# ============================================================
# Iterar sobre las filas
# ============================================================

# values devuelve un ndarray con todas las filas.
for fila in frame.values:
    print(fila)

print(type(frame.values[0]))
````

````python
# ============================================================
# Recorrer cada elemento del DataFrame
# ============================================================

# Primero se recorre cada fila y luego cada valor.
for fila in frame.values:
    for valor in fila:
        print(valor)
````

# 4.3 Indexación y slicing con DataFrames

Pandas ofrece dos formas principales para acceder a los datos de un DataFrame:

- **iloc[]**: acceso mediante posiciones numéricas.
- **loc[]**: acceso utilizando los nombres de filas y columnas.

Estas herramientas permiten seleccionar filas, columnas o regiones completas del DataFrame de forma sencilla y eficiente.

````python
# ============================================================
# Crear un DataFrame para los ejemplos de indexación
# ============================================================

d1 = {"ciudad": "Valencia", "temperatura": 10, "o2": 1}
d2 = {"ciudad": "Barcelona", "temperatura": 8}
d3 = {"ciudad": "Valencia", "temperatura": 9}
d4 = {"ciudad": "Madrid", "temperatura": 10, "humedad": 80}
d5 = {"ciudad": "Sevilla", "temperatura": 15, "humedad": 50, "co2": 6}
d6 = {"ciudad": "Valencia", "temperatura": 10, "humedad": 90, "co2": 10}

# Crear una lista de diccionarios.
lista_datos = [d1, d2, d3, d4, d5, d6]

# Crear el DataFrame indicando índices personalizados.
df_data = pd.DataFrame(lista_datos, index=list("abcdef"))

display(df_data)
````

````python
# ============================================================
# Acceder a datos utilizando iloc y loc
# ============================================================

# Acceso por posición.
print(df_data.iloc[1, 1])

# Primeras filas y columnas utilizando posiciones.
display(df_data.iloc[:3, :4])

# Acceso por etiquetas.
display(df_data.loc["a", "temperatura"])

display(df_data.loc[:"c", :"o2"])

display(df_data.loc[:"c", "temperatura":"o2"])

display(df_data.loc[:, ["ciudad", "o2"]])
````

````python
# ============================================================
# Acceso a columnas
# ============================================================

# Una columna devuelve una Serie.
print(df_data["ciudad"])

# Varias columnas devuelven un DataFrame.
display(df_data[["ciudad", "o2"]])
````

````python
# ============================================================
# Intentar acceder utilizando un entero
# ============================================================

# Esto produce un KeyError porque Pandas interpreta el 0
# como el nombre de una columna y no como la primera fila.

df_data[0]
````

````python
# ============================================================
# Obtener la primera fila utilizando iloc
# ============================================================

print(df_data.iloc[0])
````

````python
# ============================================================
# Acceder utilizando el índice semántico
# ============================================================

display(df_data.loc["a"])

display(df_data.loc[:"b"])
````

````python
# ============================================================
# Seleccionar varias filas y columnas
# ============================================================

display(
    df_data.loc[:"b"].loc[:, ["o2", "humedad"]]
)
````

````python
# ============================================================
# Diferencia entre referencia y copia
# ============================================================

display(df_data)

# Obtener una referencia de una fila.
serie = df_data.loc["a"]

print(serie)

# Modificar la referencia también modifica el DataFrame.
serie.iloc[2] = 3000

display(df_data)

# Crear una copia independiente.
df_copia = df_data.loc["a"].copy()

df_copia.iloc[2] = 9999

display(df_copia)

display(df_data)
````

````python
# ============================================================
# Acceso utilizando axis
# ============================================================

display(df_data.loc(axis=1)["ciudad"])
````

````python
# ============================================================
# Diferencia entre loc e iloc
# ============================================================

frame = pd.DataFrame(
    {
        "Name": ["Carlos", "Pedro"],
        "Age": [34, 22]
    },
    index=[1, 0]
)

display(frame)

print("Primera fila\n")
print(frame.iloc[0])

print("\nElemento cuyo índice es 0\n")
print(frame.loc[0])
````

### Observaciones

- **iloc[]** trabaja con posiciones numéricas.
- **loc[]** trabaja con los nombres de los índices y columnas.
- Al seleccionar una parte del DataFrame normalmente se obtiene una **referencia**, por lo que los cambios pueden afectar al DataFrame original.
- Si se desea modificar los datos sin alterar el DataFrame original, es recomendable utilizar el método `.copy()`.
- Cuando existe ambigüedad entre índices numéricos y posiciones, siempre es preferible utilizar `loc` o `iloc` explícitamente.

## 4.4 Objeto `Index` de Pandas

Los objetos **Index** representan los índices utilizados por las Series y los DataFrames.

### Características principales

- Permiten identificar de forma única filas o columnas.
- Son estructuras optimizadas para realizar búsquedas rápidas.
- Son **inmutables**, es decir, una vez creados no pueden modificarse directamente.
- Pueden utilizarse al crear un DataFrame personalizado.

````python
# ============================================================
# Construcción de un objeto Index
# ============================================================

import pandas as pd

# Crear un objeto Index con varios valores enteros.
ind = pd.Index([2, 3, 5, 23, 26])

# Mostrar el cuarto elemento (posición 3).
print(ind[3])

# Mostrar elementos desde el inicio tomando uno cada dos posiciones.
print(ind[::2])
````

````python
# ============================================================
# Utilizar un objeto Index al crear un DataFrame
# ============================================================

# Crear un DataFrame utilizando el objeto Index como índice.
frame = pd.DataFrame(
    {
        "Name": ["Carlos", "Pedro", "Manolo", "Luis", "Alberto"],
        "Age": [34, 22, 15, 55, 23]
    },
    index=ind
)

display(frame)
````

````python
# ============================================================
# Los objetos Index son inmutables
# ============================================================

try:
    # Intentar modificar un elemento del índice.
    ind[3] = 8
except TypeError as e:
    print("Error:", e)
````

````python
# ============================================================
# Cambiar la columna utilizada como índice
# ============================================================

# Crear nuevamente el DataFrame.
frame = pd.DataFrame(
    {
        "Name": ["Carlos", "Pedro", "Manolo", "Luis", "Alberto"],
        "Age": [34, 22, 15, 55, 23]
    },
    index=ind
)

print("DataFrame original:")
display(frame)

# Convertir la columna 'Age' en el nuevo índice.
frame.set_index("Age", inplace=True)

print("DataFrame con Age como índice:")
display(frame)
````

# 4.5 Slicing

El **slicing** permite seleccionar una parte del DataFrame.

Se puede realizar mediante:

- Filas.
- Columnas.
- Índices semánticos (`loc`).
- Índices posicionales (`iloc`).
- Condiciones lógicas.

````python
# ============================================================
# Crear un DataFrame para los ejemplos
# ============================================================

d_and_d_characters = {
    "Name": ["bundenth", "theorin", "barlok"],
    "Strength": [10, 12, 19],
    "Wisdom": [20, 13, 6]
}

character_data = pd.DataFrame(
    d_and_d_characters,
    index=["a", "b", "c"]
)

display(character_data)

# Mostrar todas las filas excepto la última.
display(character_data[:-1])

# Mostrar únicamente la segunda fila.
display(character_data[1:2])
````

````python
# ============================================================
# Seleccionar columnas específicas
# ============================================================

display(character_data[["Name", "Wisdom"]])
````

````python
# ============================================================
# Slicing utilizando iloc y loc
# ============================================================

# iloc trabaja con posiciones.
display(character_data.iloc[1:])

# loc trabaja con nombres de índices y columnas.
display(character_data.loc[:'b', 'Name':'Strength'])
````

````python
# ============================================================
# Obtener únicamente las columnas Name y Strength
# ============================================================

display(character_data.loc[:, "Name":"Strength"])
````

````python
# ============================================================
# Seleccionar filas y columnas específicas con loc
# ============================================================

display(
    character_data.loc[
        ["a", "c"],
        ["Name", "Wisdom"]
    ]
)
````

````python
# ============================================================
# Seleccionar filas y columnas específicas con iloc
# ============================================================

# Filas 0 y 2, columnas 0 y 2.
display(character_data.iloc[[0, 2], [0, 2]])

# También es posible utilizar índices negativos.
display(character_data.iloc[[0, -1], [0, -1]])
````

````python
# ============================================================
# Filtrar personajes con Strength mayor que 11
# ============================================================

display(
    character_data.loc[
        character_data["Strength"] > 11,
        ["Name", "Strength"]
    ]
)
````

````python
# ============================================================
# Filtrar personajes con Strength > 15 o Wisdom > 15
# ============================================================

display(
    character_data.loc[
        (character_data["Strength"] > 15) |
        (character_data["Wisdom"] > 15)
    ]
)
````

# 5 Cargar y guardar datos en Pandas

Pandas permite guardar y cargar datos desde distintos formatos de archivo, siendo **CSV** uno de los más utilizados.

Entre las principales funciones se encuentran:

- `to_csv()`: guarda un DataFrame en un archivo CSV.
- `read_csv()`: carga un archivo CSV dentro de un DataFrame.

Además, Pandas puede trabajar con otros formatos como:

- JSON
- Excel
- HDF5
- Parquet

Esto facilita el intercambio y almacenamiento de información para su posterior análisis.

````python
# ============================================================
# Guardar un DataFrame en un archivo CSV
# ============================================================

import os

# Crear la ruta donde se almacenará el archivo.
ruta = os.path.join("res", "o_d_d_characters.csv")

# Guardar el DataFrame en formato CSV.
# El separador puede modificarse mediante el parámetro sep.
# En este caso se utiliza ';' como separador.
character_data.to_csv(ruta, sep=';')

print("Archivo CSV guardado correctamente.")
````

````python
# ============================================================
# Leer un archivo CSV
# ============================================================

# Leer el archivo utilizando el mismo separador con el que fue guardado.
loaded = pd.read_csv(ruta, sep=';')

# Mostrar el contenido del DataFrame cargado.
display(loaded)
````

## Argumentos útiles de `to_csv()`

Al guardar un DataFrame, algunos parámetros útiles son:

- `sep`: define el separador de columnas.
- `index`: indica si se guarda el índice (`True` por defecto).
- `na_rep`: reemplaza los valores `NaN` por un texto específico.
- `encoding`: define la codificación del archivo (por ejemplo, `"utf-8"`).

### Ejemplo

```python
df.to_csv("datos.csv", sep=";", na_rep="Sin dato")
```

---

## Argumentos útiles de `read_csv()`

Al cargar un archivo CSV, algunos parámetros frecuentes son:

- `sep`: especifica el separador de columnas.
- `index_col`: selecciona la columna que actuará como índice.
- `na_values`: indica qué valores deben interpretarse como `NaN`.
- `encoding`: especifica la codificación del archivo.

### Ejemplo

```python
df = pd.read_csv(
    "datos.csv",
    sep=";",
    na_values="Sin dato"
)
```

---

Pandas también proporciona funciones para trabajar con otros formatos de almacenamiento como:

- JSON
- Excel
- HDF5
- Parquet
- SQL

Estos formatos permiten intercambiar información entre diferentes aplicaciones y proyectos de análisis de datos.

## Modificación incorrecta de un DataFrame

El siguiente ejemplo produce el conocido aviso **SettingWithCopyWarning**.

Se incluye únicamente con fines didácticos para mostrar una práctica que debe evitarse.

````python
# ============================================================
# Ejemplo que genera SettingWithCopyWarning
# ============================================================

users_dataset = pd.read_csv(
    ruta_users,
    sep="::",
    index_col=0,
    header=None,
    names=["UserID", "Gender", "Age", "Occupation", "Zip-code"],
    engine="python"
)

# Convertir la columna Age para permitir valores NaN.
users_dataset["Age"] = users_dataset["Age"].astype(float)

under_age = users_dataset[
    users_dataset["Age"] == 1
]

# Esta operación genera el SettingWithCopyWarning.
under_age.loc[:, "Age"] = np.nan

display(under_age.head())
````

````python
# ============================================================
# Solución utilizando copy()
# ============================================================

users_dataset = pd.read_csv(
    ruta_users,
    sep="::",
    index_col=0,
    header=None,
    names=["UserID","Gender","Age","Occupation","Zip-code"],
    engine="python"
)

under_age = users_dataset[
    users_dataset["Age"] == 1
]

under_age_copy = under_age.copy()

under_age_copy["Age"] = np.nan

display(under_age_copy.head())

users_dataset.loc[
    users_dataset["Age"] == 1
] = under_age_copy

display(users_dataset.head())
````

````python
# ============================================================
# Eliminar definitivamente los registros con edades inválidas
# ============================================================

users_dataset = pd.read_csv(
    ruta_users,
    sep="::",
    index_col=0,
    header=None,
    names=["UserID","Gender","Age","Occupation","Zip-code"],
    engine="python"
)

# Reemplazar las edades incorrectas por NaN.
users_dataset.loc[
    users_dataset["Age"] == 1,
    "Age"
] = np.nan

display(
    users_dataset.loc[
        pd.isnull(users_dataset["Age"])
    ].head(4)
)

# Eliminar los registros con NaN.
users_dataset.drop(
    users_dataset[
        pd.isnull(users_dataset["Age"])
    ].index,
    inplace=True
)

display(users_dataset.head())
````

````python
# ============================================================
# Agrupar usuarios por género y obtener estadísticas
# ============================================================

display(
    users_dataset.groupby(
        by="Gender"
    ).describe()
)
````

````python
# ============================================================
# Guardar el DataFrame procesado
# ============================================================

ruta_output = os.path.join(
    "res",
    "ml-1m",
    "o_users_processed.csv"
)

users_dataset.to_csv(
    ruta_output,
    sep=",",
    na_rep="null"
)

print("Archivo guardado correctamente.")
````

# 7 Ejercicios

En esta sección se realiza un pequeño análisis sobre el conjunto de datos **MovieLens 1M**, utilizando el archivo **ratings.dat**.

Los objetivos son:

- Cargar el archivo de valoraciones.
- Comprobar si existen usuarios que no hayan realizado ninguna valoración.
- Identificar cuántos usuarios tienen menos de 30 reseñas registradas.

````python
# ============================================================
# Cargar el archivo ratings.dat
# ============================================================

import os
import pandas as pd

# Construir la ruta del archivo.
ruta_ratings = os.path.join("res", "ml-1m", "ratings.dat")

# Leer el archivo indicando que utiliza "::" como separador
# y asignando nombres a las columnas.
ratings_dataset = pd.read_csv(
    ruta_ratings,
    sep="::",
    header=None,
    names=[
        "UserID",
        "MovieID",
        "Rating",
        "Timestamp"
    ],
    engine="python"
)

# Mostrar el contenido del DataFrame.
display(ratings_dataset)
````

````python
# ============================================================
# Verificar si existen usuarios sin ninguna valoración
# ============================================================

# Contar cuántos usuarios aparecen en el archivo de ratings.
n_users_with_reviews = len(
    ratings_dataset.groupby(by="UserID")
)

print("Usuarios con valoraciones:", n_users_with_reviews)

# Número total de usuarios del dataset users.dat.
n_users = len(users_dataset)

print("Total de usuarios:", n_users)

# Calcular cuántos usuarios no realizaron ninguna valoración.
print(
    "Usuarios sin valoraciones:",
    n_users - n_users_with_reviews
)
````

````python
# ============================================================
# Usuarios con menos de 30 valoraciones
# ============================================================

# Agrupar las valoraciones por usuario.
# Posteriormente se filtran únicamente aquellos usuarios
# que tienen menos de 30 reseñas registradas.
ratings_under_30 = (
    ratings_dataset
    .groupby("UserID")["UserID"]
    .filter(lambda x: len(x) < 30)
    .value_counts()
)

# Mostrar el resultado.
display(ratings_under_30)

# Mostrar cuántos usuarios cumplen esta condición.
print(
    "Usuarios con menos de 30 reviews:",
    len(ratings_under_30)
)
````
