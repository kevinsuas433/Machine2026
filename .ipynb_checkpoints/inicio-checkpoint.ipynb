3.0.2 ¿Qué es un intérprete?
</div>

<div style="

line-height:1.8;
">

<ul>

<li>Un programa que ejecuta otros programas.</li>

<li>Debe estar instalado en tu ordenador para poder ejecutar código Python.</li>

<li>
Para la especificación de
<a href="https://docs.python.org/3/reference/index.html" target="_blank">
Python
</a>,
existen varias implementaciones:
    <ul style="list-style-type:none;">
        <li>– CPython (implementación en C). Es el más común.</li>
        <li>– Jython (implementación en Java).</li>
        <li>– IronPython (implementación en .NET).</li>
    </ul>
</li>

</ul>

3.0.8 REPL
</div>

<div style="
font-family:'Times New Roman', serif;
font-size:18px;
line-height:1.7;
">

Sistema interactivo para comunicarse con el ordenador en un lenguaje, Python.

Se debe cumplir:

<ul style="list-style-type:none;">
<li>* <b>Read</b>. El ordenador pueda leer unidades como entrada.</li>
<li>* <b>Evaluate</b>. El código pueda ser procesado.</li>
<li>* <b>Print</b>. Los resultados puedan verse.</li>
<li>* <b>Loop</b>. Continuar con la conversación.</li>
</ul>

</div>

</div>

<div style="

">

<ul>

<li>Permite almacenar programas.</li>

Ficheros de texto con instrucciones Python.

        <li>– No olvidar shebang en Linux → <code>#!/usr/bin/env python3</code></li>

Terminología (varía según fuentes):
    <ul style="list-style-type:none;">

        <li>– Módulos: ficheros importados desde otros ficheros.</li>
    </ul>
</li>

<li>Se puede lanzar pasando nombre de fichero a comando python.</li>

</ul>

3.0.12 Performance Sample, NumPy vs Plain Python
</div>

````python
import numpy as np

my_arr = np.arange(10000000)  # Utilizando NumPy Arrays
my_list = list(range(10000000))  # Utilizando Listas

print(my_list)
````

````python
%time for _ in range(10): my_arr2 = my_arr * 2
%time for _ in range(10): my_list2 = [x * 2 for x in my_list]
````
