# 1 Acceso a ficheros

## 1.1 Ficheros

Los **ficheros** (*files*) son objetos predefinidos (*built-in*) de Python que permiten almacenar información de forma permanente y acceder a ella desde un programa.

Algunas de sus características más importantes son:

- Los ficheros son objetos incorporados en Python (*built-in*).
- Permiten leer y escribir información almacenada en el disco.
- No pertenecen a los tipos de datos numéricos, secuencias o diccionarios, por lo que su manipulación se realiza mediante métodos específicos.
- Para trabajar con un fichero primero es necesario abrirlo mediante la función **`open()`**.

### Abrir un fichero

La función **`open()`** permite crear un objeto de tipo fichero para poder leer o escribir información.

Su sintaxis general es la siguiente:

```python
archivo = open(nombre_archivo, modo)
```

Donde:

- **`nombre_archivo`** corresponde al nombre o la ruta del fichero.
- **`modo`** indica la forma en que se abrirá el archivo (lectura, escritura, agregar información, etc.).

Si no se especifica el modo, Python abre el archivo en **modo lectura (`r`)** de forma predeterminada.

### Consideraciones importantes

Al trabajar con archivos en Python es importante tener en cuenta los siguientes aspectos:

- Toda la información leída desde un fichero se obtiene inicialmente como una cadena de texto (**`str`**).
- Del mismo modo, al escribir datos en un fichero, Python almacena texto.
- Una vez finalizado el trabajo con el archivo, es recomendable cerrarlo utilizando el método **`close()`**, ya que esto libera los recursos utilizados por el sistema operativo y evita posibles errores.

## Lectura de archivos

Una vez que un fichero ha sido abierto mediante la función **`open()`**, Python ofrece diferentes métodos para leer su contenido. Dependiendo de la necesidad, es posible leer todo el archivo de una sola vez, recorrerlo línea por línea o almacenar su contenido dentro de una lista.

### Lectura completa de un archivo con `read()`

El método **`read()`** permite obtener todo el contenido del fichero como una única cadena de texto. Es recomendable utilizar este método cuando el archivo es pequeño y se desea trabajar con todo su contenido al mismo tiempo.

````python
# ==========================================
# Lectura completa de un archivo con read()
# ==========================================

# Abrimos el archivo en modo lectura.
mi_fichero = open("res/multiple_lines.txt")

# Leemos todo el contenido del archivo.
contenido = mi_fichero.read()

# Mostramos el contenido por pantalla.
print(contenido)

# Cerramos el archivo para liberar recursos.
mi_fichero.close()
````

### Lectura línea por línea utilizando un ciclo `for`

Los objetos de tipo fichero son iterables, por lo que pueden recorrerse mediante un ciclo **`for`**. Este método es muy útil cuando se trabaja con archivos grandes, ya que evita cargar todo el contenido en memoria.

````python
# ==========================================
# Lectura línea por línea
# ==========================================

# Abrimos el archivo.
mi_fichero = open("res/multiple_lines.txt")

# Recorremos cada línea del fichero.
for linea in mi_fichero:

    # Mostramos la línea leída.
    print(linea)

# Cerramos el archivo.
mi_fichero.close()
````

### Leer todas las líneas mediante `readlines()`

El método **`readlines()`** devuelve una lista donde cada elemento corresponde a una línea del archivo. Esto facilita recorrer posteriormente el contenido utilizando listas de Python.

````python
# ==========================================
# Leer un archivo como una lista
# ==========================================

# Abrimos el archivo.
mi_fichero = open("res/multiple_lines.txt")

# Leemos todas las líneas y las almacenamos
# dentro de una lista.
lineas = mi_fichero.readlines()

# Cerramos el archivo.
mi_fichero.close()

# Mostramos la lista obtenida.
print(lineas)
````

### Construcción de rutas independientes del sistema operativo

Los sistemas operativos utilizan separadores de rutas diferentes (`/` o `\`). Para evitar problemas de compatibilidad, Python proporciona el módulo **`os`**, cuyo método **`os.path.join()`** construye rutas válidas para cualquier plataforma.

````python
# ==========================================
# Construcción de rutas con os.path.join()
# ==========================================

# Importamos el módulo os.
import os

# Construimos la ruta del archivo.
ruta = os.path.join("res", "multiple_lines.txt")

# Mostramos la ruta generada.
print(ruta)
````

También es posible construir una ruta utilizando una lista de directorios y desempaquetándola con el operador `*`.

````python
# ==========================================
# Construcción de rutas usando una lista
# ==========================================

# Lista con los componentes de la ruta.
dataset_file_path = [
    "res",
    "multiple_lines.txt"
]

# Construimos la ruta desempaquetando la lista.
ruta = os.path.join(*dataset_file_path)

# Mostramos el resultado.
print(ruta)
````

### Uso de la sentencia `with`

La forma más recomendada de trabajar con archivos en Python consiste en utilizar la sentencia **`with`**.

Esta estructura crea un **context manager**, el cual garantiza que el archivo será cerrado automáticamente al finalizar el bloque, incluso si ocurre algún error durante la ejecución.

````python
# ==========================================
# Lectura utilizando with
# ==========================================

# Abrimos el archivo utilizando un context manager.
with open(ruta) as mi_fichero:

    # Recorremos el archivo línea por línea.
    for linea in mi_fichero:

        # end="" evita agregar una línea en blanco extra.
        print(linea, end="")
````

### Abrir varios archivos simultáneamente

La sentencia **`with`** también permite abrir varios archivos dentro del mismo bloque de código, facilitando su lectura y asegurando que todos sean cerrados automáticamente al finalizar.

````python
# ==========================================
# Abrir varios archivos con with
# ==========================================

# Construimos las rutas de ambos archivos.
ruta1 = os.path.join("res", "one_line.txt")
ruta2 = os.path.join("res", "multiple_lines.txt")

# Abrimos ambos archivos simultáneamente.
with open(ruta1) as fichero1, open(ruta2) as fichero2:

    # Mostramos el contenido del primer archivo.
    print(fichero1.readlines())

    print()

    # Mostramos el contenido del segundo archivo.
    print(fichero2.readlines())
````

# 1.2 Modos de acceso

Al abrir un archivo mediante la función **`open()`**, es posible indicar el modo en el que se accederá al fichero. El modo de apertura determina si el archivo será utilizado para lectura, escritura o ambas operaciones.

Los modos más utilizados se muestran en la siguiente tabla.

| Modo | Descripción |
|:----:|-------------|
| `r` | Solo lectura. Es el modo predeterminado. |
| `w` | Solo escritura. Si el archivo existe, su contenido se elimina antes de escribir. |
| `x` | Crea un nuevo archivo. Produce un error si el archivo ya existe. |
| `a` | Abre el archivo para agregar información al final. Si no existe, lo crea. |
| `r+` | Permite leer y escribir sobre el mismo archivo. |
| `b` | Acceso en modo binario. Se combina con otros modos (`rb`, `wb`, etc.). |
| `t` | Acceso en modo texto. Es el modo utilizado por defecto. |

# 1.3 Acceso para escritura

Además de leer información, Python permite crear y modificar archivos utilizando distintos modos de escritura.

En el siguiente ejemplo se generará un archivo de texto y posteriormente se escribirán varias líneas utilizando los métodos **`write()`** y **`writelines()`**.

````python
# ==========================================
# Crear un archivo y escribir información
# ==========================================

# Importamos el módulo para trabajar con rutas.
import os

# Función que genera una lista de cadenas.
def crear_lista(tamanyo):

    # Creamos una lista vacía.
    lista = []

    # Agregamos los números convertidos a texto.
    for i in range(tamanyo):

        # Cada elemento termina con un salto de línea.
        lista.append(str(i) + "\n")

    # Devolvemos la lista creada.
    return lista


# Construimos la ruta del archivo.
ruta = os.path.join("res", "a_dummy.txt")

# Abrimos el archivo en modo escritura.
with open(ruta, "wt") as fichero:

    # Escribimos una línea de encabezado.
    fichero.write("Cabecera del ciclo del for en el ejemplo\n")

    # Generamos la lista.
    lista = crear_lista(10)

    # Escribimos todas las líneas de la lista.
    fichero.writelines(lista)

# Confirmamos que el archivo fue creado.
print("Archivo creado correctamente.")
````

## Buffering

Cuando un programa escribe información en un archivo, los datos **no siempre se almacenan inmediatamente en el disco**.

Python utiliza un **buffer**, que es una zona de memoria temporal donde se guardan los datos antes de escribirlos definitivamente en el archivo.

El contenido del buffer se guarda en el disco cuando ocurre alguna de las siguientes acciones:

- Se cierra el archivo mediante **`close()`**.
- Se utiliza la sentencia **`with`**, que lo cierra automáticamente.
- Se invoca el método **`flush()`**, que fuerza la escritura inmediata del contenido del buffer.

````python
# ==========================================
# Ejemplo del uso del buffer
# ==========================================

# Construimos la ruta del archivo.
ruta = os.path.join("res", "FicheroParaEscritura.txt")

# Abrimos el archivo para escritura.
fichero_write = open(ruta, "w")

# Escribimos una línea.
fichero_write.write("footer")

# Abrimos el mismo archivo para lectura.
fichero_read = open(ruta, "r")

print("Contenido antes de flush():")
print(fichero_read.readlines())

# Forzamos la escritura del buffer al disco.
fichero_write.flush()

# Regresamos al inicio del archivo para volver a leerlo.
fichero_read.seek(0)

print()

print("Contenido después de flush():")
print(fichero_read.readlines())

# Cerramos ambos archivos.
fichero_write.close()
fichero_read.close()
````

### Explicación del ejemplo

En este ejemplo se observa el funcionamiento del **buffer**.

1. Se abre un archivo para escritura y se almacena el texto `"footer"`.
2. Antes de ejecutar **`flush()`**, la información puede seguir permaneciendo únicamente en el buffer.
3. Al llamar a **`flush()`**, el contenido se escribe físicamente en el archivo.
4. Finalmente, el archivo se vuelve a leer para comprobar que la información ya está disponible.

### Conclusión

Al trabajar con archivos es recomendable utilizar la sentencia **`with`**, ya que administra automáticamente la apertura y el cierre de los ficheros, evitando pérdidas de información y liberando correctamente los recursos del sistema.

Cuando se utilizan archivos abiertos manualmente mediante **`open()`**, es importante recordar cerrar el archivo con **`close()`** o utilizar **`flush()`** cuando se necesite escribir inmediatamente el contenido almacenado en el buffer.

# 1.4 Archivos CSV

Los archivos **CSV** (*Comma-Separated Values*) son uno de los formatos más utilizados para almacenar y compartir datos tabulares.

Cada línea del archivo representa un registro y los campos se encuentran separados por un delimitador, que normalmente es una **coma (,)**, aunque también pueden utilizarse otros caracteres como punto y coma (`;`) o barra vertical (`|`).

Este formato es ampliamente utilizado en áreas como **Ciencia de Datos**, **Machine Learning**, **Inteligencia Artificial** y **Análisis de Datos**, debido a su simplicidad y compatibilidad con numerosas herramientas.

### Lectura de un archivo CSV

Python incorpora el módulo **`csv`**, el cual facilita la lectura de archivos CSV sin necesidad de procesar manualmente cada línea del fichero.

En el siguiente ejemplo se leerá un archivo denominado **`tabla_operaciones.csv`** y se mostrarán sus columnas por pantalla.

````python
# ==========================================
# Lectura de un archivo CSV
# ==========================================

# Importamos los módulos necesarios.
import os
import csv

# Construimos la ruta del archivo.
ruta = os.path.join("res", "tabla_operaciones.csv")

# Abrimos el archivo utilizando un context manager.
with open(ruta) as fichero:

    # Creamos un lector CSV indicando
    # que los datos están separados por comas.
    data_reader = csv.reader(fichero, delimiter=",")

    # Recorremos cada fila del archivo.
    for linea in data_reader:

        # Mostramos las dos primeras columnas.
        print(linea[0] + " ---- " + linea[1])
````

### Escritura de un archivo CSV

Además de leer archivos CSV, Python también permite generar nuevos archivos utilizando el módulo **`csv`**.

En el siguiente ejemplo se leerá un archivo CSV separado por comas y posteriormente se generará una copia utilizando como separador el carácter **`|`**.

````python
# ==========================================
# Escritura de un archivo CSV
# ==========================================

# Construimos la ruta del archivo original.
ruta = os.path.join("res", "tabla_operaciones.csv")

# Abrimos el archivo para lectura.
f = open(ruta)

# Creamos el lector CSV.
data_reader = csv.reader(f, delimiter=",")

# Ruta del nuevo archivo.
ruta_o = os.path.join("res", "4_tabla_operaciones.csv")

# Abrimos el nuevo archivo para escritura.
with open(ruta_o, "w", newline="") as f2:

    # Creamos el escritor CSV utilizando
    # el carácter "|" como separador.
    csv_writer = csv.writer(f2, delimiter="|")

    # Recorremos todas las filas.
    for linea in data_reader:

        # Mostramos la fila por pantalla.
        print(linea)

        # Escribimos la fila en el nuevo archivo.
        csv_writer.writerow(linea)

# Cerramos el archivo original.
f.close()
````

### Archivo generado

Después de ejecutar el programa se creará un nuevo archivo llamado **`4_tabla_operaciones.csv`** dentro de la carpeta **`res`**.

Su contenido tendrá un formato similar al siguiente:

```text
Operacion|Descripcion
a + b|suma a y b
a - b|resta a menos b
a / b|a dividido
a // b|a dividido entre b (quitando decimales)
a % b|devuelve el resto de la división a/b (modulus)
a * b|a multiplicado por b
a ** b|a elevado a b
```

### Conclusión

El módulo **`csv`** simplifica el procesamiento de archivos tabulares en Python. Gracias a las clases **`csv.reader()`** y **`csv.writer()`**, es posible leer, modificar y generar archivos CSV de forma sencilla, manteniendo la compatibilidad con herramientas ampliamente utilizadas como Microsoft Excel, LibreOffice Calc y diversas bibliotecas de análisis de datos como **Pandas**.

# 2 Docstrings

Los **docstrings** (*documentation strings*) son cadenas de texto utilizadas para documentar módulos, funciones, clases y otros objetos en Python.

Su objetivo es describir el funcionamiento del código para facilitar su comprensión, mantenimiento y reutilización.

Python almacena automáticamente esta documentación en el atributo especial **`__doc__`**, permitiendo consultarla durante la ejecución del programa.

### Consultar la documentación de un objeto

Python incorpora la función **`help()`**, la cual permite visualizar la documentación asociada a un objeto, incluyendo funciones, clases, módulos y métodos.

Por ejemplo, es posible consultar la documentación de la función incorporada **`open()`**.

````python
# ==========================================
# Consultar la documentación de una función
# ==========================================

# Mostramos la ayuda de la función open().
help(open)
````

### Crear un docstring

No es necesario modificar manualmente el atributo **`__doc__`**.

Para documentar una función basta con escribir una cadena de texto entre **comillas triples (`"""`)** inmediatamente después de la definición de la función y antes del código ejecutable.

````python
# ==========================================
# Ejemplo de un docstring
# ==========================================

# Definimos una función.
def funcion_de_prueba():
    """
    Esta es la documentación de la función de prueba.
    """

    # La función no realiza ninguna acción.
    pass


# Mostramos la documentación utilizando help().
help(funcion_de_prueba)
````

### Acceder al atributo `__doc__`

Si únicamente se desea obtener el texto del docstring, puede consultarse directamente mediante el atributo especial **`__doc__`**.

````python
# ==========================================
# Mostrar únicamente el docstring
# ==========================================

# Imprimimos el contenido del atributo __doc__.
print(funcion_de_prueba.__doc__)
````

### Diferencia entre `help()` y `__doc__`

Aunque ambos permiten acceder a la documentación, presentan diferencias importantes:

- **`help()`** muestra información adicional sobre el objeto, como su firma, descripción y otros detalles útiles.
- **`__doc__`** devuelve únicamente el texto del docstring definido por el programador.

### Documentación de módulos

La función **`help()`** también puede utilizarse para consultar la documentación completa de un módulo de Python.

En el siguiente ejemplo se visualizará la documentación del módulo **`sys`**.

````python
# ==========================================
# Consultar la documentación de un módulo
# ==========================================

# Importamos el módulo sys.
import sys

# Mostramos su documentación.
help(sys)
````

### Estructura recomendada de un docstring

Cuando un docstring ocupa varias líneas, normalmente sigue la siguiente estructura:

- Un resumen breve en la primera línea.
- Una línea en blanco.
- Una descripción más detallada.
- Otra línea en blanco antes del código de la función.

Este formato facilita la lectura tanto para los desarrolladores como para las herramientas de documentación automática.

````python
# ==========================================
# Ejemplo de un docstring multilínea
# ==========================================

# Definimos una función con documentación extensa.
def funcion_de_prueba2():
    """
    Esta es la línea para el resumen.

    Este es el párrafo donde se puede escribir una descripción
    más detallada de la función. Observa que el resumen y la
    descripción van separados por una línea en blanco.

    También existe otra línea en blanco antes de comenzar
    el código de la función.
    """

    # La función no realiza ninguna acción.
    pass


# Mostramos la documentación generada.
help(funcion_de_prueba2)
````

### Conclusión

Los **docstrings** constituyen una buena práctica de programación, ya que permiten documentar el propósito y funcionamiento del código directamente dentro del programa.

Además de mejorar la legibilidad, facilitan el mantenimiento del software y permiten que herramientas como **`help()`** o los generadores automáticos de documentación obtengan información detallada sobre módulos, funciones y clases.

### Comentarios y Docstrings

Aunque tanto los comentarios como los docstrings sirven para documentar el código, cada uno tiene un propósito diferente.

- Los **comentarios** (`#`) suelen utilizarse para explicar instrucciones individuales, expresiones o pequeños bloques de código.
- Los **docstrings** están pensados para documentar elementos de mayor nivel, como módulos, funciones, clases y métodos.
- Los docstrings pueden consultarse durante la ejecución del programa mediante la función **`help()`** o el atributo **`__doc__`**, mientras que los comentarios únicamente sirven como ayuda para el programador y son ignorados por Python durante la ejecución.

## 2.1 Sphinx

**Sphinx** es una herramienta especializada en la generación automática de documentación para proyectos desarrollados en Python.

Su principal ventaja es que puede generar documentación profesional a partir de los docstrings escritos dentro del código fuente.

Entre sus características más importantes se encuentran:

- Generación automática de documentación.
- Ideal para proyectos de mediana y gran escala.
- Compatible con múltiples formatos de salida, entre ellos:
  - HTML
  - PDF (mediante LaTeX)
  - ePub
  - Manuales en línea
- Integración con plataformas de documentación y repositorios de código.

### ¿Por qué utilizar Sphinx?

Cuando un proyecto crece y contiene decenas o cientos de módulos, mantener una documentación manual resulta complicado.

Sphinx permite generar automáticamente páginas de documentación utilizando la información almacenada en los **docstrings**, evitando duplicar esfuerzos y manteniendo la documentación sincronizada con el código.

## 2.2 Estilo

Es recomendable que todos los docstrings de un proyecto mantengan un formato uniforme.

Cuando varias personas trabajan sobre el mismo código, seguir un mismo estilo facilita la lectura, el mantenimiento y la generación automática de documentación.

Por esta razón, los equipos de desarrollo suelen definir una guía de estilo antes de comenzar un proyecto.

### Google Python Style Guide

Una de las guías de estilo más utilizadas en proyectos profesionales es la **Google Python Style Guide**.

Esta guía establece recomendaciones sobre cómo documentar correctamente los diferentes componentes de un programa, entre ellos:

- Módulos.
- Funciones y métodos.
- Clases.
- Parámetros.
- Valores de retorno.
- Excepciones generadas.

Seguir una guía de estilo permite que la documentación sea consistente y fácil de comprender para cualquier desarrollador que participe en el proyecto.

### Conclusión

Una buena documentación es tan importante como un buen código.

El uso adecuado de **docstrings**, junto con herramientas como **Sphinx** y una guía de estilo consistente, facilita el mantenimiento de los programas, mejora la colaboración entre desarrolladores y permite generar documentación profesional de forma automática.

# 2.3 Ejercicios

A continuación se presentan algunos ejercicios para practicar la lectura y escritura de archivos de texto, así como el procesamiento de información almacenada en ficheros utilizando Python.

## Ejercicio 1

**Enunciado**

Escribe una función que reciba la ruta de un fichero de texto y una cadena de caracteres, y determine si dicha cadena aparece dentro del archivo.

````python
# ==========================================
# Ejercicio 1
# Buscar una cadena dentro de un archivo
# ==========================================

# Importamos el módulo os para construir rutas.
import os

# Definimos la función.
def buscar_cadena(ruta, cadena):

    # Abrimos el archivo en modo lectura.
    with open(ruta, "r", encoding="utf-8") as fichero:

        # Leemos todo el contenido del archivo.
        contenido = fichero.read()

    # Verificamos si la cadena existe.
    return cadena in contenido


# Ruta del archivo de prueba.
ruta = os.path.join("res", "multiple_lines.txt")

# Probamos la función.
print(buscar_cadena(ruta, "Python"))
print(buscar_cadena(ruta, "Java"))
````

## Ejercicio 2

**Enunciado**

Escribe una función que reciba una lista, una ruta de destino y un número **n**. La función debe crear un archivo que contenga únicamente los primeros **n** elementos de la lista, controlando adecuadamente los valores fuera de rango.

````python
# ==========================================
# Ejercicio 2
# Guardar los primeros n elementos
# ==========================================

# Definimos la función.
def guardar_elementos(lista, ruta, n):

    # Verificamos que n sea válido.
    if n < 0 or n > len(lista):
        print("Error: valor de n fuera de rango.")
        return

    # Creamos el archivo.
    with open(ruta, "w", encoding="utf-8") as fichero:

        # Escribimos únicamente los primeros n elementos.
        for elemento in lista[:n]:
            fichero.write(str(elemento) + "\n")


# Lista de ejemplo.
datos = ["Ana", "Luis", "Pedro", "María", "José"]

# Ruta del archivo.
ruta = os.path.join("res", "primeros_elementos.txt")

# Ejecutamos la función.
guardar_elementos(datos, ruta, 3)

print("Archivo generado correctamente.")
````

## Ejercicio 3

**Enunciado**

Escribe una función que reciba la ruta de un fichero de texto y devuelva un diccionario con la frecuencia de aparición de cada palabra contenida en el archivo.

````python
# ==========================================
# Ejercicio 3
# Contar frecuencia de palabras
# ==========================================

# Definimos la función.
def frecuencia_palabras(ruta):

    # Abrimos el archivo.
    with open(ruta, "r", encoding="utf-8") as fichero:

        # Leemos el contenido.
        texto = fichero.read()

    # Dividimos el texto en palabras.
    palabras = texto.split()

    # Diccionario donde almacenaremos las frecuencias.
    frecuencias = {}

    # Recorremos todas las palabras.
    for palabra in palabras:

        # Si la palabra ya existe, incrementamos el contador.
        if palabra in frecuencias:
            frecuencias[palabra] += 1

        # Si aparece por primera vez, iniciamos el contador.
        else:
            frecuencias[palabra] = 1

    # Devolvemos el diccionario.
    return frecuencias


# Ruta del archivo de ejemplo.
ruta = os.path.join("res", "primeros_elementos.txt")

# Mostramos el resultado.
print(frecuencia_palabras(ruta))
````
